# Theory

A coded aperture imaging system produces an image from a series of exposures, that have been modulated with a mask. This produces measurements in *mask-space*, but we desire an image in *pixel-space*. 
The decoder is a $2n^2$ network, that converts between mask-space and pixel space. 

```python
from IPython.display import SVG
SVG('../img/decoder.svg')
```

The vector coded aperture system (`vecap`) being developed here  will combine the imaging abilities of a coded aperture techniques with vector measurments. Essentially, this requires a calibration and DUT measurements to be performed for each mask pattern. Once this is done, a corrected measurements for each mask are summed together using the decoder to produce an image in pixel space.

# `vecap.decoder`  Test

```python
from  vecap.decoder import Decoder 
import skrf as rf
from skrf.media  import Freespace    

    
# create a freespace Calibration template 
delta = 40
freq =rf.N('/home/alex/Dropbox/Vector Coded Aperture Imaging Data/Contrast Test/0/ds,0.s1p').frequency
air = Freespace(frequency = freq, z0=50)
ideals = [ air.delay_short(k*delta,'um',name='ds,%i'%k) for k in range(5)]
cal_template = rf.OnePort(measured = ideals, ideals = ideals, 
                 sloppy_input=True, is_reciprocal=False)

c=Decoder(base_dir='/home/alex/Dropbox/Vector Coded Aperture Imaging Data/Contrast Test/',
                cal_template = cal_template)
```
