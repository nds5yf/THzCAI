```python
import numpy as np
import matplotlib.pyplot as plt
import math
step = 0.5
arr = np.array(key.schottky)
plt.imshow(arr)
plt.colorbar().set_label(label = 'Voltage')
plt.ylabel('Y (' + str(step) + ' mm)')
plt.xlabel('X (' + str(step) + ' mm)')
plt.title('ZBD Voltage vs Position')
xpos = 0
ypos = 0
x80 = 0
y80 = 0
radius = 0
resolution = 20
for x in range(0, key.resolution):
    for y in range(0, key.resolution):
        if float(key.schottky[x][y]) >= float(np.amax(arr)):
            ypos = x
            xpos = y
        if (float(np.amax(arr)) * 0.75) < float(key.schottky[x][y]) < (0.85 * float(np.amax(arr))):
            y80 = x
            x80 = y
radius = math.sqrt(math.pow((x80 - xpos), 2) + math.pow((y80 - ypos), 2)) * step
plotstr = 'Max: ' + str(np.amax(arr)) + ' V' + '\n' + 'Min: ' + str(np.amin(arr)) + ' V' + '\n' + 'Center: ' + '(' + str(xpos) + ', ' + str(ypos) + ')' + '\n' + 'Est. Radius: ' + str(radius) + ' mm'
plt.text(0, 0, plotstr, fontsize=10, verticalalignment='top',bbox=dict(facecolor='white', alpha=1))
fig = plt.gcf()
fig.gca().add_artist(plt.Circle((xpos,ypos),radius * 2,color='w', alpha=1, fill = False))
```
